## Задание 1.
Реализовать Наивный алгоритм возведения в степень. 
Сравнить его результаты и скорость вычислений с функцией PowerMod. Сколько арифметических операций выполняется при работе данного алгоритма? Является ли он полиномиальным?

```julia
function pow(a, d, n)
    b = BigInt(1)
    for i = 1:d
        b = (a * b) % n
    end

    return Int(b)
end
```

## Задание 3.
Реализовать Расширенный алгоритм Евклида. Сравнить его значения и скорость со встроенной функцией ExtendedGCD. Сколько арифметических операций выполняется при работе данного алгоритма? Является ли он полиномиальным?

```julia
function extGCD(a::Unsigned, b::Unsigned)
    M = [1 0; 0 1]
    a, b = Int(a), Int(b)
    q, r = divrem(a, b)

    while r != 0
        a = b
        b = r
        M *= [0 1; 1 -q]

        q, r = divrem(a, b)
    end

    return (Int(M[1,2]), Int(M[2,2]))
end
```

```julia
@time extGCD(Unsigned(7), Unsigned(3))
```

```julia
@time gcdx(7,3)[[2,3]]
```

## Задание 4.
Найти обратные относительно умножения к элементам a, b в ℤn. Вариант соответствует номеру по порядку в списке группы.

```julia
n = BigInt(42530430997171493050900585519445269701954006270353944787367883)
a = BigInt(949014432282168334172)
b = BigInt(2847043296846505002513)
```

```julia
function invZ(a::Number, n::Number)
    g, x, y = gcdx(a, n);
    if (g != 1)
        println("Oops");
    else
        x = (x % n + n) % n;
        return x;
    end
end
```

```julia
invZ(a,n)
```

```julia
invZ(b,n)
```

```julia
invmod(a,n)
```

```julia
invmod(b,n)
```

## Задание 5.
Реализовать переполнение памяти умножением,используя функцию Memory-Constrained. Быстро ли работает данный алгоритм? Сколько арифметических операций выполняет данный алгоритм? Чему равна T(N)? Является ли он полиномиальным?

```julia
function strangeMult(n::Number)
    a = BigInt(2);
    for i = 1:(floor(log2(n))+1)
        a = a * a;
    end

    return a
end
```

```julia
@time strangeMult(1000000000)
```
